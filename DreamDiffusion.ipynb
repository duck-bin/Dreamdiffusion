```python
from torch.utils.data import Dataset
import numpy as np
import glob
import os

class eeg_pretrain_dataset(Dataset):
    def __init__(self, path, desired_length, transform=None):
        self.input_paths = glob.glob(os.path.join(path, '*.npy'))
        assert len(self.input_paths) != 0, 'No data found in the specified path.'
        self.transform = transform
        self.desired_length = desired_length  # 통일할 길이 L

    def __len__(self):
        return len(self.input_paths)

    def __getitem__(self, idx):
        # 1. 데이터 로드
        eeg_array = np.load(self.input_paths[idx])  # shape: (128, L_i)

        # 2. 길이 통일 (패딩 또는 자르기)
        current_length = eeg_array.shape[1]
        if current_length >= self.desired_length:
            # 앞에서부터 desired_length만큼 자르기
            eeg_array = eeg_array[:, :self.desired_length]
        else:
            # 뒤에 0으로 패딩 추가
            pad_width = ((0, 0), (0, self.desired_length - current_length))
            eeg_array = np.pad(eeg_array, pad_width, mode='constant')

        # 3. 필요하면 transform 적용
        if self.transform:
            eeg_array = self.transform(eeg_array)

        # 4. NumPy 배열을 텐서로 변환
        eeg_tensor = torch.from_numpy(eeg_array).float()  # Float 텐서로 변환

        return eeg_tensor
```

```python
import os
import sys
import torch
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader

# 프로젝트의 코드 디렉토리를 Python 경로에 추가
sys.path.append('/content/dreamdiffusion/code')

# 필요한 클래스 임포트
from sc_mbm.mae_for_eeg import MAEforEEG
from dataset import eeg_pretrain_dataset
```

```python
# 모델 초기화 (학습 시 사용한 하이퍼파라미터와 동일하게 설정)
model = MAEforEEG(
    time_len=512,
    patch_size=4,
    embed_dim=1024,
    decoder_embed_dim=512,
    depth=24,
    num_heads=16,
    decoder_num_heads=16,
    mlp_ratio=1.0,
    focus_range=None,
    focus_rate=0.6,
    img_recon_weight=0.5,
    use_nature_img_loss=False
)

# 디바이스 설정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# 저장된 모델 가중치 로드
checkpoint_path = '/content/dreamdiffusion/pretrains/eeg_pretrain/checkpoint.pth'
model.load_state_dict(torch.load(checkpoint_path, map_location=device))
model.eval()
```

```python
# 데이터셋 로드
dataset = eeg_pretrain_dataset(
    path='/content/dreamdiffusion/datasets/mne_data/',
    roi='VC',
    patch_size=4,
    transform=None,
    aug_times=1,
    num_sub_limit=None,
    include_kam=True,
    include_hcp=True
)
```

```python
@torch.no_grad()
def plot_recon_figures(model, device, dataset, num_figures=5, mask_ratio=0.75):
    dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
    model.eval()
    fig, axs = plt.subplots(num_figures, 3, figsize=(15, 5 * num_figures))

    for i in range(num_figures):
        sample = next(iter(dataloader))['eeg']
        sample = sample.to(device)
        _, pred, mask = model(sample, mask_ratio=mask_ratio)

        sample_with_mask = sample.cpu().squeeze(0)[0].numpy().reshape(-1, model.patch_size)
        pred = model.unpatchify(pred).cpu().squeeze(0)[0].numpy()
        sample = sample.cpu().squeeze(0)[0].numpy()
        mask = mask.cpu().numpy().reshape(-1)

        cor = np.corrcoef(pred, sample)[0,1]

        x_axis = np.arange(sample.shape[-1])

        # Ground Truth
        axs[i, 0].plot(x_axis, sample)
        axs[i, 0].set_title('Ground Truth')

        # Masked Ground Truth
        s = 0
        for x, m in zip(sample_with_mask, mask):
            if m == 0:
                axs[i, 1].plot(x_axis[s:s+len(x)], x)
            s += len(x)
        axs[i, 1].set_title('Masked Ground Truth')

        # Reconstruction
        axs[i, 2].plot(x_axis, pred)
        axs[i, 2].set_title(f'Reconstruction (cor={cor:.4f})')

    plt.tight_layout()
    plt.show()
```

```python
# 플롯 함수 실행
plot_recon_figures(model, device, dataset, num_figures=5, mask_ratio=0.75)
```
